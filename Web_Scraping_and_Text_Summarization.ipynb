```python
import scrapy
from bs4 import BeautifulSoup
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize

# Scrapy spider for Wikipedia scraping
class WikipediaSpider(scrapy.Spider):
    name = "wikipedia"
    start_urls = [
        "https://en.wikipedia.org/wiki/Alexander_the_Great"
    ]

    # Define parsing logic
    def parse(self, response):
        soup = BeautifulSoup(response.text, "html.parser")
        headings_and_content = []

        # Find all headings and their corresponding content
        for heading in soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"]):
            if heading.text.strip():
                section_data = {
                    "heading": heading.text.strip(),
                    "content": "",
                }
                next_element = heading.find_next_sibling()
                while next_element and next_element.name not in ["h1", "h2", "h3", "h4", "h5", "h6"]:
                    section_data["content"] += next_element.text + " "
                    next_element = next_element.find_next_sibling()
                headings_and_content.append(section_data)

        # Summarize the content under each heading
        for section_data in headings_and_content:
            heading = section_data["heading"]
            content = section_data["content"]

            # Tokenize the content into sentences
            sentences = sent_tokenize(content)

            # Calculate word frequency to identify important words
            words = word_tokenize(content)
            words = [word.lower() for word in words if word.isalpha()]
            stop_words = set(stopwords.words("english"))
            words = [word for word in words if word not in stop_words]
            freq_dist = FreqDist(words)

            # Select sentences based on word frequency
            selected_sentences = []
            for sentence in sentences:
                if any(word in sentence.lower() for word in freq_dist.keys() if freq_dist[word] > 1):
                    selected_sentences.append(sentence)

            # Print the heading and summarized content
            print(heading)
            print(" ".join(selected_sentences))
            print("\n")

# Run the Scrapy spider
if __name__ == "__main__":
    from scrapy.crawler import CrawlerProcess
    process = CrawlerProcess()
    process.crawl(WikipediaSpider)
    process.start()
```

```python
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize

# Define the URL of the Wikipedia page to scrape
url = "https://en.wikipedia.org/wiki/Alexander_the_Great"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find all the headings and their corresponding content
sections = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])
headings_and_content = []

for section in sections:
    if section.text.strip():  # Ignore empty headings
        section_data = {
            "heading": section.text.strip(),
            "content": "",
        }
        next_element = section.find_next_sibling()
        while next_element and next_element.name not in ["h1", "h2", "h3", "h4", "h5", "h6"]:
            section_data["content"] += next_element.text + "\n"
            next_element = next_element.find_next_sibling()
        headings_and_content.append(section_data)

# Set the desired number of sentences in the summary
sentence_limit = 2

# Summarize the content under each heading
for section_data in headings_and_content:
    heading = section_data["heading"]
    content = section_data["content"]

    # Tokenize the content into sentences
    sentences = sent_tokenize(content)

    # Calculate word frequency to identify important words
    words = word_tokenize(content)
    words = [word.lower() for word in words if word.isalpha()]
    stop_words = set(stopwords.words("english"))
    words = [word for word in words if word not in stop_words]
    freq_dist = FreqDist(words)

    # Select a limited number of sentences based on word frequency
    selected_sentences = []
    for sentence in sentences:
        if any(word in sentence.lower() for word in freq_dist.keys() if freq_dist[word] > 1):
            selected_sentences.append(sentence)
            if len(selected_sentences) >= sentence_limit:
                break

    # Print the heading and summarized content
    print(heading)
    print(" ".join(selected_sentences))
    print("\n")
```
